```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset directly from seaborn
iris_data = sns.load_dataset('iris')

# Print a success message
print("Libraries imported and dataset loaded successfully!")
```

```python
# 1. Print the shape of the dataset (number of rows, number of columns)
print("Dataset Shape:", iris_data.shape)

# 2. Print the column names to see what features we are working with
print("\nColumn Names:", iris_data.columns.tolist())

# 3. Display the first 5 rows of the dataset
iris_data.head()
```

```python
# Print a concise summary of the dataset (data types, missing values)
print("--- Dataset Info ---")
iris_data.info()

# Print descriptive statistics (mean, min, max, etc.)
print("\n--- Descriptive Statistics ---")
iris_data.describe()
```

```python
# 1. Scatter Plot: Showing relationships between features
plt.figure(figsize=(8, 6)) # Sets the size of the graph

# We use seaborn (sns) to create a scatter plot. 
# We'll plot petal_length vs petal_width, and color code ('hue') by species.
sns.scatterplot(data=iris_data, x='petal_length', y='petal_width', hue='species', s=100)

plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.show()
```

```python
# 2. Histograms: Showing value distributions for all numerical features
# The 'bins' parameter decides how many bars to divide the data into
iris_data.hist(figsize=(10, 8), bins=15, edgecolor='black')

# Add a main title above all the subplots
plt.suptitle("Histograms of Iris Features", fontsize=16)

# Display the plots
plt.show()
```

```python
# 3. Box Plots: Identifying outliers in the dataset
plt.figure(figsize=(10, 6))

# Create a boxplot for all numerical columns in the dataset
sns.boxplot(data=iris_data)

plt.title("Box Plots of Iris Features (Identifying Outliers)", fontsize=14)
plt.ylabel("Centimeters")
plt.show()
```

Scatter Plot: The scatter plot made it really clear that setosa flowers are easy to identify. Their petals are a lot smaller than the other two species, so all their data points clump together in the bottom left corner.

Histograms: The sepal width data looks like a normal, even bell curve. But for the petal lengths and widths, there's a big empty gap in the middle of the charts. This makes sense because the tiny setosa flowers are separated from the rest.

Box Plots: Using the box plots was a great way to spot outliers. I noticed a few stray dots sitting just outside the whiskers specifically for the sepal_width column, which means a couple of those flowers were unusually wide or narrow.
