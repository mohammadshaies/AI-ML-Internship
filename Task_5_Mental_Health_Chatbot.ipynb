After successfully loading and tokenizing a subset of the Facebook 'Empathetic Dialogues' dataset, we fine-tuned a pre-trained DistilGPT2 Large Language Model locally.

**Key Findings:**
* The model successfully learned the structural formatting of the training data (recognizing the `User (emotion): ... \nBot:` prompt pattern).
